### *Image Generator*

```python
input = "" #@param {type:"string"}
seek = "" #@param {type:"string"}
filter = "" #@param {type:"string"}

if not seek:
  if not filter:
    !ffmpeg -loglevel error -y -i "{input}" -frames:v 1 -q 0 i.jpg
  else:
    !ffmpeg -loglevel error -y -i "{input}" -vf "{filter}" -frames:v 1 -q 0 i.jpg
else:
  if not filter:
    !ffmpeg -loglevel error -y -ss {seek} -i "{input}" -frames:v 1 -q 0 i.jpg
  else:
    !ffmpeg -loglevel error -y -ss {seek} -i "{input}" -vf "{filter}" -frames:v 1 -q 0 i.jpg

import glob
from IPython.display import Image, display
jpg_files = glob.glob('*.jpg') + glob.glob('*.jpeg')
for file in jpg_files:
    display(Image(file))
```

### *Trimmer*

```python
input = "" #@param {type:"string"}
output = "" #@param {type:"string"}
start = "" #@param {type:"string"}
end = "" #@param {type:"string"}
re_encode = False #@param {type:"boolean"}
from datetime import datetime
x = datetime.strptime(start, "%H:%M:%S")
y = datetime.strptime(end, "%H:%M:%S")
end = y - x
if not re_encode:
  !ffmpeg -y -loglevel error -i "$input" -ss "$start" -t "$end" -c copy "$output"
else:
  !ffmpeg -y -loglevel error -ss "$start" -i "$input" -to "$end" -crf 0 -c:a copy "$output"
```
